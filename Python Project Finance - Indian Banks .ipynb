```python
#We will predicting the future stock prices for each constituent using Monte Carlo Method. 
```

```python
rets = bank_list.dropna()
#dropna() is to drop any null values in the list
```

```python
rets
```

```python
days = 365 

dt = 1/days

mu = rets.mean()['YESBANK']

sigma = rets.std()['YESBANK']

def stock_monte_carlo (start_price, days, mu, sigma):
    
    price = np.zeros(days)
    price[0] = start_price
    
    shock = np.zeros(days)
    
    drift = np.zeros(days)
    
    for x in range(1, days):
        
        shock[x] = np.random.normal(loc= mu*dt, scale = sigma*np.sqrt(dt))
        
        drift[x] = mu*dt
        
        price[x]= price[x-1] + (price[x-1]* (drift[x] + shock[x]))
        
    return price
```

```python
start_price = 185.899
```

```python

for run in range(100):
    plt.plot(stock_monte_carlo (start_price, days, mu, sigma))
    
plt.xlabel ('Days')
plt.ylabel ('Price')
plt.title ('Monte Carlo Analysis for YESBANK')
```
